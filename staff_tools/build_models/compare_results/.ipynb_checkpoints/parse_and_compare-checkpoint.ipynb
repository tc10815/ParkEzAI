```python
import os
import json

# Load the JSON data
with open('labels.json', 'r') as json_file:
    test_labels = json.load(json_file)
```

```python
with open('django_classifications', 'r') as file:
    text = file.read()

django_labels = {}
for x in text.split('Image Name: camfeeds/coldwatermi/')[1:]: 
    dictionary = json.loads(x[43:-1])
    django_labels[x[0:28]] = dictionary
```

```python
total_wrong = 0
total_right = 0

for key in django_labels.keys():
    print(key)
    for spot in django_labels[key]:
        print(f'\t{spot}')
        print(f'\t\t{test_labels[key][spot]}')
        print(f'\t\t{django_labels[key][spot]}')
        print()
        if test_labels[key][spot] == django_labels[key][spot]:
            total_right += 1
        else:
            total_wrong += 1
```
